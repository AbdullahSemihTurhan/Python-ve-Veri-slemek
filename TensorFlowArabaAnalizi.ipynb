# model olusturma ve egitmeye baslama


```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=10)#veri alırken ki alış algoritması
```

```python
len(x_train) # kac adet train icin ayirdi
```

```python
len(x_test)# kac adet test icin ayirdi
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scalerobje=MinMaxScaler()
```

```python
x_train=scalerobje.fit_transform(x_train)
```

```python
x_test=scalerobje.fit_transform(x_test)
```

```python
from tensorflow.keras.models import Sequential #nöron olusturur
from tensorflow.keras.layers import Dense#katmanolusturur
```

```python
#dense = nöronlarin birbirine bagli oldugu bir katman 
```

```python
x_train.shape #9090 veri , 5 cesit(ozellik oldugunu göruruz)
```

```python
model=Sequential() #modeli olusturduk
```

```python
model.add(Dense(12,activation="relu"))
model.add(Dense(12,activation="relu"))
model.add(Dense(12,activation="relu"))
model.add(Dense(12,activation="relu"))
model.add(Dense(1))
model.compile(optimizer="adam",loss="mse")
```

```python
model.fit(x=x_train,y=y_train,validation_data=(x_test,y_test),batch_size=250 ,epochs=300)
#batch_size=250 = modele butun veriyi ayni anda vermek yorar o yuzden 250 ser veriyoruz
```

```python
kayipVerisi=pd.DataFrame(model.history.history)
```

```python
kayipVerisi.head()
```

```python
kayipVerisi.plot()
```

```python
#val_loss = dogrulama kaybı
#belli bir yerden sonra val loss ile loss ayrı yerlere gidiyorsa sıkıntı vardır
#ancak oldukca yaklaşırlarsa doğru yoldasındadır demektir
```
