# verileri bolmek

```python
y=dataFrame["Type"].values
x=dataFrame.drop("Type",axis=1).values #type kolonunu dusurcez
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.33,random_state=15)
```

```python
from sklearn.preprocessing import MinMaxScaler # veriyi daraltma uyarlama 
```

```python
scaler=MinMaxScaler()
```

```python
scaler.fit(x_train)
```

```python
x_train=scaler.transform(x_train)
```

```python
x_test=scaler.transform(x_test)
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Activation,Dropout
from tensorflow.keras.callbacks import EarlyStopping
```

# EARLYSTOPPİN VE DROPOUT

```python
#OverFitting Problemlerini Cozmek icin kullanilir
#EarlyStopping
#gitgide birbirine yaklasması , hatalarin azalmis olması gerekir
#valloss saçma sapan artisa gecerse Loss u durdurmak icin EarlyStopping kullancaz
```

```python
x_train.shape
```

```python
model=Sequential()

model.add(Dense(units=30,activation="relu"))
#30 tane icine noron koy #kolon sayisina esit olmasi onerilir

model.add(Dense(units=15,activation="relu"))
model.add(Dense(units=15,activation="relu"))
model.add(Dense(units=1,activation="sigmoid")) # sigmoid = 0 ile 1 arasinda deger verir

model.compile(loss="binary_crossentropy",optimizer="adam")
```

```python
model.fit(x=x_train,y=y_train,epochs=600,validation_data=(x_test,y_test),verbose=1)
#epochs neye gore belirlenir veri setin uzerinden kac kere gecip train , ogreneceğinin yazariz
#fazla yazarsan buga girer
#asagida sonucu var
#600 kere datalarin uzerinden geciyor
#validationdata=dogrulama verisini direk model.fit icersine vekmek icin,sonradan ugrasmamak icin
```

```python
model.history.history #losslar gozukur
```

```python
modelkaybi=pd.DataFrame(model.history.history)
```

```python
modelkaybi.plot() 
```

```python
#gitgide birbirine yaklasması , hatalarin azalmis olması gerekir
#valloss saçma sapan artisa gecerse Loss u durdurmak icin EarlyStopping kullancaz
```

```python
model=Sequential()

model.add(Dense(units=30,activation="relu"))
#30 tane icine noron koy #kolon sayisina esit olmasi onerilir

model.add(Dense(units=15,activation="relu"))
model.add(Dense(units=15,activation="relu"))
model.add(Dense(units=1,activation="sigmoid")) # sigmoid = 0 ile 1 arasinda deger verir

model.compile(loss="binary_crossentropy",optimizer="adam")
```

```python
earlyStopping=EarlyStopping(monitor="val_loss",mode="min",verbose=1,patience=25) #patience=sabir
#monitor val_loss u takip edecek ve Hata olusmasını engelleyecek
```

```python
model.fit(x=x_train,y=y_train,epochs=600,validation_data=(x_test,y_test),verbose=1,callbacks=[earlyStopping])
#callbacks e [dizi icersinde ] earlystopping yaziyoruz , ukarda tanimladigimizi,
```

```python
#246 da otomatik durdurdu 
```

```python
modelKayip=pd.DataFrame(model.history.history)
```

```python
modelKayip.plot()
```

MEMNUN DEGİL İSEK DROPOUT YAPARIZ

```python
model=Sequential()

model.add(Dense(units=30,activation="relu"))
model.add(Dropout(0.6))

model.add(Dense(units=15,activation="relu"))
model.add(Dropout(0.6))

model.add(Dense(units=15,activation="relu"))
model.add(Dropout(0.6))

model.add(Dense(units=1,activation="sigmoid")) 
#son(cıktı) katmanı dısında her katmana ustteki gibi konur
model.compile(loss="binary_crossentropy",optimizer="adam")
```

```python
model.fit(x=x_train,y=y_train,epochs=600,validation_data=(x_test,y_test),verbose=1,callbacks=[earlyStopping])
#callbacks e [dizi icersinde ] earlystopping yaziyoruz , ukarda tanimladigimizi,
```

```python
kayipDataFrame=pd.DataFrame(model.history.history)
```

```python
kayipDataFrame.plot()
```

```python
#DropOut Kullaninca zigzag olması çok normal , full çözum olamaz ancak çok daha iyi olur
```

```python
tahminlerim=model.predict_classes(x_test)  #siniflari tahmin et,girdilere gore yeni tahminler cıkartir
```

```python
tahminlerim
```

# TAHMİNLERİMİZİ DUZGUN DEGERLENDİRMEK İCİN

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,tahminlerim)) #kiyasliyor
```

```python
print(confusion_matrix(y_test,tahminlerim))
#ilk dizenin 2. ogesi yanlıs,geri kalanlari dogru bulmus demektir
```
